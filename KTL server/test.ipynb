```python
import time, threading
import xml.etree.ElementTree as ET

VERBOSE = False

def parse_xml(xmlfile):
    tree = ET.parse(xmlfile)
    root = tree.getroot()

    keyword_names = []
    keyword_types = []
    capability_types = []

    for keyword in root.findall('keyword'):
        name = keyword.find('name').text
        keyword_type = keyword.find('type').text
        capability_type = keyword.find('capability').text #.get('type')

        print('Parsed keyword: name=%s\ttype=%s\tcapability=%s' % (name, keyword_type, capability_type))
        capability_types.append(capability_type)
        keyword_names.append(name)
        keyword_types.append(keyword_type)

    return keyword_names, keyword_types, capability_types
```

```python
keyword_names, keyword_types, capability_types = parse_xml('LFC.xml.sin')
```
