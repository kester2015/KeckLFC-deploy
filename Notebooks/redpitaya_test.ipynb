```python
from LFC.Hardware import *
import numpy as np
fg = FG('rp-f0602c.local')
rp_fc = RedPitaya('rp-f072a9.local')
fg.connect()
rp_fc.connect()
fin = lambda x: np.mean(rp_fc.acquire(port = 1))
fout = lambda y: rp_fc.dc(port = 2, offset = y)
fc_lock = Servo(fin, fout, name = 'Filter Cavity lock')
```

```python
rp_fc.startosc(port = 1)
rp_fc.setosc(dec = 2**10, xlim = None, ylim = (-10,10), trigger = "NOW")
```

```python
rp_fc.stoposc()
```

```python
fg.startfg(phase = 0)
rp_fc.ramp(port=2)
```

```python
fg.stopfg()
rp_fc.disable(port=2)
```

```python
fc_lock.tune(P=1e-3, I =0.01, dt = 0)
fc_lock.start(3,0.3,[-1,1])
```

```python
fc_lock.monitor = True
```
