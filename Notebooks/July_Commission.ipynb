```python
import matplotlib.pyplot as plt
import numpy as np

n_pt = 35
x_freq = np.linspace(192.3-n_pt*16e-3, 192.3+n_pt*16e-3, 16)
y_phase = ((x_freq-192.3) ** 2 * (2*np.pi) ** 2 * 1.3/2*-2.75) + 0/((x_freq-192.3) ** 4 * (2*np.pi) ** 2 * 1.3/2+1)
plt.plot(x_freq, y_phase,label='original')

y_phase[5:11] *= 1
#y_phase[:3] *= 1



plt.plot(x_freq, y_phase,label='after')
plt.legend()
plt.show()


ws.set_interp_phase(x_freq, y_phase)
ws.setBandPass(span=5)
# ws.set2ndDisper(d2=-2.8)
ws.plotStatus()
ws.writeProfile()
ws.printStatus()
```

```python
filename = r"C:\Users\HSFLFC\Desktop\Keck\Logs\NIRSPEC_0718\NIRSPEC_Test_2023_0718_1.csv"
import time
with  open(filename,"w") as f:
    #TODO: write table head
    f.write(time.strftime('%Z')+",")


    f.write("Pritel Cur (mA)"+",")

    f.write("Pritel Pwr (mW)"+",")
    f.write("Rb lock error"+",")

    f.write("Rb lock measure"+",")

    f.write("Rb lock output"+",")

    f.write("Freq Counter (Hz)"+",")

    f.write("IM lock error"+",")
    f.write("IM lock measure"+",")
    f.write("IM lock output"+",")

    f.write("PPLN Temp (C)"+",")
    f.write("WG Temp (C)"+",")

    f.write("\n")

ii = 0
while True:
    try:
        with open(filename,"a") as f:
            TSTAMP = time.ctime()

            f.write(TSTAMP+",")

            f.write(f"{ptamp.pwrAmp:.5f}"+",")
            f.write(f"{ptamp.outputPwr_mW:.5f}"+",")

            f.write(f"{servo_RB.amplified_error:.8f}"+",")
            f.write(f"{servo_RB.measure_input:.8f}"+",")
            f.write(f"{servo_RB.output_voltage:.8f}"+",")

            f.write(f"{cnt90.measFreq('c'):.5f}"+",")

            f.write(f"{servo_IM.amplified_error:.8f}"+",")
            f.write(f"{servo_IM.measure_input:.8f}"+",")
            f.write(f"{servo_IM.output_voltage:.8f}"+",")

            f.write(f"{tec_ppln.get_temp():.5f}"+",")
            f.write(f"{tec_wg.get_temp():.5f}"+",")

            f.write("\n")

            ii = ii+1
            osa.save_trace('a',r"C:\Users\HSFLFC\Desktop\Keck\Logs\NIRSPEC_0718\Minicomb" +"\\" +str(ii)+ ".mat", plot=False)
            
    except Exception as e:
        print(e)
    time.sleep(20)
```
